**8. Dropping irrelevent columns from the data to predict the Churn**

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
data.head()
```

**9. Display the unique values in each column**

```python
# Display unique values in each column
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Unique values in {column}: {unique_values}")
```

**10. Encoding Categorical Data**

```python
data = pd.get_dummies(data, drop_first = True)
```

```python
data.head()
```

**11. Check the Data Balance**

```python
data['Exited'].value_counts()
```

```python
import seaborn as sns
sns.countplot(data=data, x='Exited')
```

```python
X = data.drop(['Exited'], axis=1)
y = data['Exited']
```

**12. Handling the Data Imbalance**

```python
 from imblearn.over_sampling import SMOTE
 X_res, y_res = SMOTE().fit_resample(X,y)
 y_res.value_counts()
```

**13. Spliting the Data into Training and Testing set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res, test_size=0.2, random_state = 42)
```

**14. Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**15. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test, y_pred1)
```

```python
precision_score(y_test, y_pred1)
```

```python
recall_score(y_test, y_pred1)
```

```python
f1_score(y_test, y_pred1)
```

**16. SVC**

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(X_train, y_train)
```

```python
y_pred2 = svm.predict(X_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
precision_score(y_test, y_pred2)
```

**17. KNeighbors Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred3 = knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred3)
```

```python
precision_score(y_test, y_pred3)
```

```python
recall_score(y_test, y_pred3)
```

**18. Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred4 = dt.predict(X_test)
```

```python
accuracy_score(y_test, y_pred4)
```

```python
precision_score(y_test, y_pred4)
```

**19. Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred5 = rf.predict(X_test)
accuracy_score(y_test, y_pred5)
```

```python
X_train.shape
```

```python
precision_score(y_test, y_pred5)
```

```python
recall_score(y_test, y_pred5)
```

**20. Gradient Boosting Classifier**

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred6 = gb.predict(X_test)
accuracy_score(y_test, y_pred6)
```

```python
precision_score(y_test, y_pred6)
```

**21. To Display the final result in each ML model**

```python
final_data = pd.DataFrame({'Models': ['LR','SVC', 'KNN', 'DT', 'RF', 'GBC'], 'ACC': [accuracy_score(y_test, y_pred1), accuracy_score(y_test, y_pred2), accuracy_score(y_test, y_pred3), accuracy_score(y_test, y_pred4), accuracy_score(y_test, y_pred5), accuracy_score(y_test, y_pred6)], 'PRE': [precision_score(y_test, y_pred1), precision_score(y_test, y_pred2), precision_score(y_test, y_pred3), precision_score(y_test, y_pred4), precision_score(y_test, y_pred5), precision_score(y_test, y_pred6)]})
```

```python
final_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot
sns.barplot(x=final_data['Models'], y=final_data['ACC'])

# Add titles and labels
plt.title('Model Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy')

# Show the plot
plt.show()
```
